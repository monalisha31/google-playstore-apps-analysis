```python
#FINDING THE TOTAL CATEGORY OF APPS
import pandas as pd
import numpy as np

ds = pd.read_csv("/home/nbuser/library/assignment 1/googleplaystore.csv")
ds['Category'].unique()
```

```python
#MOST INSTALLED APPS 
import pandas as pd
import numpy as np

ds = pd.read_csv("/home/nbuser/library/assignment 1/googleplaystore.csv")
ds['Installs'] = ds.Installs.str.replace(",","")
ds['Installs'] = ds.Installs.str.replace("+","")
ds['Installs'] = ds.Installs.replace("Free",np.nan)
ds['Installs'] = ds['Installs'].astype(float)
ds.drop_duplicates(inplace=True)
ds[ds['Installs'] == 1000000000.0]
```

```python
#APPS WITH MOST REVIEWS
import pandas as pd
import numpy as np

ds = pd.read_csv("/home/nbuser/library/assignment 1/googleplaystore.csv")
ds.drop_duplicates(inplace=True)
ds[ds.Reviews>40000000]
```

```python
#MOST EXPENSIVE APP
import pandas as pd
import numpy as np

ds = pd.read_csv("/home/nbuser/library/assignment 1/googleplaystore.csv")
ds.drop_duplicates(inplace=True)
newPrice = []

for row in ds.Price:
    if row!= "0":
        newrow = float(row[1:])
    else:
        newrow = 0 
        
    newPrice.append(newrow)
        
ds.Price = newPrice
ds[ds.Price == ds.Price.max()]
```

```python
#CATEGORY GRAPH
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

ds = pd.read_csv("/home/nbuser/library/assignment 1/googleplaystore.csv")
ds.drop_duplicates(inplace=True)
ds.Category.value_counts().plot(kind='barh',figsize= (12,8))
plt.show()
```

```python
#APPS WITH FULL RATINGS
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

ds = pd.read_csv("/home/nbuser/library/assignment 1/googleplaystore.csv")
ds.drop_duplicates(inplace=True)

print("No. of Apps with full ratings: ",ds.Rating[ds['Rating'] == 5 ].count())
```

```python
#PRICE DISTRIBUTION PIE CHART
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

ds = pd.read_csv("/home/nbuser/library/assignment 1/googleplaystore.csv")


newPrice = []

for row in ds.Price:
    if row!= "0":
        newrow = float(row[1:])
    else:
        newrow = 0 
        
    newPrice.append(newrow)
        
ds.Price = newPrice

plt.pie(ds.Type.value_counts(), labels=['Free', 'Paid'], autopct='%1.1f%%')
```

```python
#ANDROID USERS DISTRIBUTION GRAPH
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 

ds = pd.read_csv("/home/nbuser/library/assignment 1/googleplaystore.csv")
ds.dropna(inplace=True)
newVer = []

for row in ds['Android Ver']:
    try:
        newrow = float(row[:2])
    except:
        newrow = 0  
    
    newVer.append(newrow)
    
ds['Android Ver'] =  newVer
sns.countplot(ds['Android Ver'])
```
